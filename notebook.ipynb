The aim of this project is to extract emotions from comments on Twitter in order to help airplanes companies to know how customers are satisfied. And what are the features that we can extract from text data?

```python
"""
We will start by reading the data using pandas librarie
"""
import pandas as pd
clean_data = pd.read_csv('Tweets.csv')
```

A first step to take a look at the data by showing the 3 first rows of it.

```python
clean_data.head(3)
```

So each tweet is defined by a set of features the most of them are just Identifiers which we will not use as  client senetiment will be expressed in **text** column. Also the real sentiment is given in **airline_sentiment** which will be our target (label).

```python
import seaborn as sns
sns.countplot(x = "airline_sentiment", data = clean_data)
```

By counting the number of sentiments for each class we can conclude that tweets with negative sentiments are more frequent. Therefore, a first finding will be that Machine learning models that we will build will likely predict negative sentiments with more accuracy. It will be a good point because companies care more about negative reviews in order to improve their services.

```python
"""
We will keep just tweets text with its sentiment
"""
data = clean_data.drop(['tweet_id', 'airline_sentiment_confidence','negativereason', 'negativereason_confidence', 'airline',
       'airline_sentiment_gold', 'name', 'negativereason_gold','retweet_count', 'tweet_coord', 'tweet_created','tweet_location',
                        'user_timezone']
                       , axis = 1)
data.head()
```

As a base line model we will use **catboost  model**. And for text encoding  **TfidfVecorizer** which a simple model that encode text data.

```python
import sklearn
from   sklearn.model_selection import train_test_split
from   catboost import CatBoostClassifier
from   sklearn.feature_extraction.text import TfidfVectorizer
import joblib
import numpy as np
```

```python
"""
X: Input text
y: Target (sentiment)
"""
X = data.text # Model input
y = data.airline_sentiment  # Model output
"""
In order to test our model , we split the data to train/test.
"""
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) # Data split train/test
"""
VECTORIZE DATA
"""
tf_transformer = TfidfVectorizer()
tf_transformer.fit(X_train)
X_train_tfidf = tf_transformer.transform(X_train)
X_test_tfidf  = tf_transformer.transform(X_test)
"""
Model training
"""
base_model = CatBoostClassifier(silent=True)
base_model.fit(X_train_tfidf, y_train)
```

Now lets check the accuracy of our model on predicting sentiments from text.

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(base_model, X_test_tfidf, y_test)  
```

As we can observe Negative class has more accurate predictions, about 2865 text tweets are negative and our model predicted it as neagtive also. As accuracy its 2865 by (2865+155+65) which is 92%.
